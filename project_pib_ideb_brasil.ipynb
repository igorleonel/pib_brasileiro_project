## Descrição do projeto

---

Neste projeto, vamos realizar uma análise e visualização dos dados do Produto Interno Bruto (PIB) dos estados brasileiros entre 2002 e 2020, assim como do Índice de Desenvolvimento da Educação Básica (IDEB) entre 2005 e 2021. O objetivo é explorar as dinâmicas econômicas e educacionais do Brasil ao longo dos anos, respondendo a questionamentos específicos por meio de visualizações de dados.
<br> <br/>
Primeiramente, analisaremos os dados do PIB, utilizando diversas técnicas de visualização, como gráficos de composição, cascata, áreas e inclinação, sempre buscando destacar as informações mais relevantes. A cada passo, os dados serão transformados em gráficos informativos para comunicar de forma clara as tendências e padrões encontrados.
<br> <br/>
> **Base de dados:**

A base de dados que será utilizada contém informações detalhadas sobre o PIB de cada estado, incluindo:

1. **ano:** ano de representação dos dados

2. **sigla_uf:** sigla do Estado representada na base de dados

3. **regiao:** região do Estado

4. **pib:** valor do PIB do Estado

5. **impostos_liquidos:** impostos líquidos do Estado

6. **va:** valor adicionado bruto a preços correntes das atividades econômicas

7. **va_agropecuaria:** valor adicionado bruto a preços correntes da agropecuária

8. **va_industria:** valor adicionado bruto a preços correntes da indústria

9. **va_servicos:** valor adicionado bruto a preços correntes dos serviços (exceto administração pública, defesa, educação e saúde públicas e seguridade social)

10. **va_adespss:** valor adicionado bruto a preços correntes da administração pública, defesa, educação, saúde pública e seguridade social
<br> <br/>

Cada questionamento levantado será respondido com o uso de visualizações apropriadas, combinando elementos visuais como cores, anotações e destaques para guiar o público aos insights mais importantes. Assim, este projeto serve como uma análise detalhada e visual das informações econômicas e educacionais do Brasil, proporcionando uma visão clara das evoluções e distribuições dos dados.

### **Pergunta 1 -** Houve uma significativa mudança na distribuição do PIB por região comparando os valores de 2002 e 2020?

#### Tratando os dados para gerar a visualização

Para responder ao primeiro questionamento, vamos utilizar o gráfico do tipo **composição estática**, mais especificamente o gráfico de pizza. Embora o gráfico de pizza seja controverso em algumas aplicações devido à dificuldade de comparar com precisão as proporções, ele ainda é útil quando queremos destacar a participação relativa de cada parte em relação ao todo, especialmente para análises de uma única variável categórica, como as regiões do Brasil.

Neste caso, o gráfico de pizza nos ajudará a visualizar as proporções do PIB distribuídas pelas regiões do Brasil em dois anos distintos (2002 e 2020). Como estamos lidando com uma divisão simples do total do PIB entre as regiões, o gráfico de pizza é uma escolha adequada para oferecer uma visão rápida e intuitiva das mudanças percentuais na distribuição.

<img src="https://github.com/igorleonel/pib_ideb_brasileiro_project/blob/main/imagens/Tipos_Graficos/Gr%C3%A1ficos%20de%20composi%C3%A7%C3%A3o%20(propor%C3%A7%C3%A3o%20dos%20dados).png?raw=True" alt="inserir alt" height="343">

<font size=2>Fonte: Andrew Abela (Extreme Presentation/Biuwer)</font>

```python
# Criando um df com os dados desejados
df_pib_2002 = df_pib.query('ano == 2002')[['regiao', 'pib']]
df_pib_2020 = df_pib.query('ano == 2020')[['regiao', 'pib']]

# Agregando os dados
df_pib_2002 = df_pib_2002.groupby('regiao').sum().sort_values(by='pib', ascending=False)
df_pib_2020 = df_pib_2020.groupby('regiao').sum().sort_values(by='pib', ascending=False)

# Visualizando as tabelas
display(df_pib_2002, df_pib_2020)
```

#### Gerando o gráfico

```python
## Configurando o gráfico com parâmetros que potencializam a visualização dos dados

# Importando as bibliotecas
import matplotlib.pyplot as plt

# Área do gráfico
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Definindo as cores do gráfico
cores = [AZUL3, LARANJA1, AZUL5 , VERDE1, CINZA5]

# Personalizando o gráfico
plt.suptitle('Produto Interno Bruto (PIB) por região', fontsize=18, color=CINZA1, ha='right', x=.5, y=1.05)

# GRÁFICO 1 - PIB de 2002 por região
ptc, text, _ = axs[0].pie(data=df_pib_2002, x='pib', labels=df_pib_2002.index, autopct='%.1f%%', pctdistance=.6, textprops={'size':12, 'fontweight': 'bold', 'color':BRANCO}, colors=cores)

# Título do gráfico 1
axs[0].annotate('Em 2002', xy=(50, 350), size=16, xycoords='axes points',
                bbox={'boxstyle': 'round', 'fc':BRANCO, 'ec':CINZA3})

# Ajustando as cores das regiões relativas as fatias
for i, p in enumerate(ptc):
  text[i].set_color(p.get_facecolor())

#######################################

# GRÁFICO 2 - PIB de 2020 por região
pct, text, _ = axs[1].pie(data=df_pib_2020, x='pib', labels=df_pib_2020.index, autopct='%1.f%%', pctdistance=0.6, textprops={'size': 12, 'fontweight': 'bold', 'color': BRANCO}, colors=cores)

# Título do Gráfico 2
axs[1].annotate('Em 2020', xy=(50, 350), size=16, xycoords='axes points',
                bbox=dict(boxstyle='round', fc=BRANCO, ec=CINZA3))

# Ajustando as cores das regiões relativas as fatias
for i, p in enumerate(pct):
  text[i].set_color(p.get_facecolor())

#######################################

axs[0].annotate('A região Sudeste foi a única que \nreduziu a participação no PIB',
                xy=(180, 245), xycoords='axes points', xytext=(280, 295), textcoords='axes points',
                bbox=dict(boxstyle='round', fc=BRANCO, ec=CINZA3),
                size=10, arrowprops=dict(arrowstyle='->', fc=CINZA1, connectionstyle='arc, angleA=-90,angleB=0,armA=30, armB=30, rad=5'))

axs[1].annotate('\n\n', xy=(140, 245), xycoords='axes points', xytext=(-50, 295), textcoords='axes points',
            size=10, arrowprops=dict(arrowstyle="->", fc=CINZA1, connectionstyle="arc,angleA=-90,angleB=0,armA=30,armB=-30,rad=5"));
```

<font color=green size=4> Conclusões: </font>

---

Aqui notamos um uso correto de gráfico de pizza que trouxemos 5 fatias no total. Tecemos uma comparação entre dois períodos separadamente, visto que o gráfico de pizza representa composições estáticas.

Podemos notar que apenas a região Sudeste recuou em participação no PIB no período que abrange 18 anos (2002 - 2020). Ou seja, foi possível notar que o PIB foi mais distribuído. Podemos perceber esse comportamento sem nem precisar analisar os dados absolutos, apenas os relativos.

### **Pergunta 4 -** Como está distribuído o PIB do Estado da Bahia em 2020, separado por impostos líquidos e os valores adicionados brutos dos bens e serviços produzidos?

Para responder à pergunta sobre a distribuição do PIB do Estado da Bahia em 2020, separado por impostos líquidos e pelos valores adicionados brutos dos bens e serviços produzidos, utilizaremos o **gráfico de cascata**. Conforme indicado na imagem abaixo, o gráfico de cascata é apropriado para visualizações de acumulação e subtração de totais, que é exatamente o que queremos representar.
<br><br/>
Nesse caso, estamos interessados em mostrar como o PIB é composto a partir da soma dos valores adicionados brutos de diferentes setores econômicos (como a agropecuária, a indústria e os serviços), além dos impostos líquidos. O gráfico de cascata nos permite visualizar o impacto de cada componente do PIB de forma clara, mostrando como cada setor e os impostos contribuem para o total do PIB da Bahia em 2020.

#### Tratando os dados para gerar a visualização

```python
# Criando um novo dataframe para visualização
bahia = df_pib.query("sigla_uf == 'BA' & ano == 2020")[['pib', 'impostos_liquidos', 'va']]

# Renomeando as colunas
bahia.rename(columns={'pib': 'PIB', 'impostos_liquidos': 'Impostos Líquidos', 'va': 'Valor Adicionado Bruto'}, inplace=True)

# Transformando as colunas em linhas
bahia = bahia.melt(var_name='Indicadores', value_name='Valores')
bahia
```

Precisamos fazer algumas pequenas alterações para construir o nosso gráfico com o novo dataframe. Observando os valores representados, percebemos que, quando eles forem adicionados ao nosso código, precisarão de uma formatação melhor.

Se representarmos esses valores do jeito que estão, será difícil visualizar se são bilhões, trilhões, enfim, quais seriam esses valores.

Além disso, precisaremos representar o imposto líquido como um valor negativo. Apesar do valor estar positivo, como queremos partir de um valor maior, o PIB, para o VA, que é menor, temos que mudar o sinal do imposto líquido.

Finalmente, criaremos outra coluna chamada Medidas, que vai indicar o que a medida corresponde. Se é uma medida de ponto inicial, de um ponto final, ou se é uma medida relativa, de acréscimo ou decréscimo.Vamos começar esses três processos!

<img src="https://github.com/igorleonel/pib_ideb_brasileiro_project/blob/main/imagens/Tipos_Graficos/Gr%C3%A1ficos%20de%20composi%C3%A7%C3%A3o%20(compondo%20dados%20estaticos%20e%20din%C3%A2micos).png?raw=True" alt="inserir alt" height="343">

<font size=2>Fonte: Andrew Abela (Extreme Presentation/Biuwer)</font>

```python
# Gerando a função para criar o gráfico de barras empilhadas

def grafico_va():
  # Importando a biblioteca
  import matplotlib.pyplot as plt

  # Área do gráfico e tema da visualização
  fig, ax = plt.subplots(figsize=(10, 7))

  # Definindo as cores do gráfico
  cores = [AZUL3, LARANJA1, CINZA5, VERDE3]

  # Gerando o gráfico de barras empilhadas 100%
  df_va_separado.plot(kind='barh', stacked=True, color=cores, ax=ax)

  ## Personalizando o gráfico
  plt.suptitle('Valores adicionados de bens e serviços do PIB brasileiro em 2020', size=18, color=CINZA1, ha='right', x=.8, y=1.01)
  plt.title('Em distribuição dos setores por regiao (%)\n', fontsize=14, color=CINZA3, pad=15, ha='right', x=0.35)
  ax.legend(bbox_to_anchor=(1, 1), bbox_transform=ax.transAxes, fontsize=10, loc='upper left',)
  ax.set_ylabel('')
  ax.set_xticklabels([])
  ax.yaxis.set_tick_params(labelsize=14, color=CINZA2)
  ax.set_frame_on(False)

  # Valores das barras
  for container in ax.containers:
    labels = [f'{valor.get_width()*100:.0f}%' for valor in container]
    ax.bar_label(container, label_type='center', labels=labels, size=10, color=CINZA2, fontweight='bold')

  return ax, cores

ax, cores = grafico_va()
```

#### **Anotações e legenda de dados**

---

Vamos trabalhar com uma função para escrita de texto utilizando diferentes cores.

> (Adaptada de https://github.com/empathy87/storytelling-with-data)



```python
# Função para escrita de texto utilizando diferentes cores. Não existe uma
# função nativa para isso. (Adaptado de https://github.com/empathy87/storytelling-with-data)
from matplotlib import transforms

def texto_colorido(x, y, texto, cores, esp=20, ax=None, **kw):
    cores = list(reversed(cores))
    t = ax.transData
    canvas = ax.figure.canvas

    for i, linha in enumerate(reversed(texto.split('\n'))):
        frases = linha.split('||')
        for s, cor in zip(frases, cores[i]):
            texto = ax.text(x, y, s, color=cor, transform=t, **kw)
            texto.draw(canvas.get_renderer())
            ex = texto.get_window_extent()
            t = transforms.offset_copy(texto._transform, x=ex.width,
                                       units='dots')

        t = transforms.offset_copy(ax.transData, x=0, y=(i + 1) * esp, units='dots')
```

```python
## Configurando o gráfico com parâmetros que potencializam a visualização dos dados

# Gerando o gráfico de barras empilhadas e salvando as cores que vamos utilizar no texto
ax, cores = grafico_va()

## Trabalhando com o texto colorido para legenda e anotações

# Legenda das barras
ax.legend().remove()
texto_colorido(
    0, 4.4, '$\\bf{Serviços}$ || | || $\\bf{Indústria}$ || | || $\\bf{ADESPSS}$ || | || $\\bf{Agropecuária}$',
    cores = [[cores[0], CINZA3, cores[1], CINZA3, cores[2], CINZA3, cores[3]]],
    ax=ax, fontsize=12
)

# Anotando uma conclusão no gráfico
texto_colorido(
    1.05, 2,                                                             # coordenadas
    'O setor de ||$\\bf{serviços}$|| corresponde a maior fatia\n'        # texto
    'do PIB em todas as regiões. Na Região ||$\\bf{Sudeste}$||,\n'
    'ele representa ||$\\bf{mais\ de\ 60\%}$|| do PIB frente a outros setores.\n'
    '\n'
    'O setor de ||$\\bf{ADESPSS\ (Administração,\ Defesa,\ Educação,}$\n'
    '$\\bf{Saúde\ Pública\ e\ Seguridade\ Social)}$|| é o ||$\\bf{2º\ mais\ relevante}$||\n'
    'no PIB da Região ||$\\bf{Nordeste}$||.\n'
    '\n'
    'O setor com menor participação é o da ||$\\bf{agropecuária}$||.\n'
    'Na Região ||$\\bf{Centro-Oeste}$|| ele praticamente empata com\n'
    'o setor da ||$\\bf{indústria}$|| em peso na região.',
    [[CINZA3, AZUL3, CINZA3],         # linha 1                          # cores
     [CINZA3, CINZA1, CINZA3],        # linha 2
     [CINZA3, CINZA1, CINZA3],        # linha 3
     [CINZA3],                        # linha 4
     [CINZA3, CINZA5],                # linha 5
     [CINZA5, CINZA3, CINZA1],        # linha 6
     [CINZA3, CINZA1, CINZA3],        # linha 7
     [CINZA3],                        # linha 8
     [CINZA3, VERDE3, CINZA3],        # linha 9
     [CINZA3, CINZA1, CINZA3],        # linha 10
     [CINZA3, LARANJA1, CINZA3]       # linha 11
    ],
    esp=22,         # espaçamento
    ax=ax,          # figura onde desenhar o texto
    fontsize=12)

fig.show()
```

Para este tipo de pergunta, iremos utilizar o gráfico de inclinação.

Com o gráfico de inclinação, é possível identificar tendências de crescimento ou declínio da participação da indústria no PIB para cada região, ressaltando de forma clara as mudanças mais significativas e os movimentos entre os períodos analisados.

```python
# Selecionando os dados desejados
anos = [2010, 2015, 2020]
df_inc = df_pib.query('ano == @anos')[['regiao', 'ano', 'va_industria']]

# Agrupando os dados da região
df_inc = pd.crosstab(index=df_inc.ano, columns=df_inc.regiao, values=df_inc.va_industria, aggfunc='sum', normalize='index')
df_inc = (df_inc*100).round(2)
df_inc = df_inc[['Sudeste', 'Sul', 'Nordeste', 'Centro-Oeste', 'Norte']]
df_inc
```

```python
# # Gerando o gráfico
def grafico_slope_pib():
  # Importando a biblioteca
  import matplotlib.pyplot as plt

  # Área do gráfico e tema da visualização
  fig, ax1 = plt.subplots(figsize=(7, 7))

  # Definindo as cores do gráfico
  cores = [AZUL3, LARANJA1, AZUL5, VERDE3, CINZA4]

  # Gerando um axis duplo partilhando o eixo x e os mesmos limites em y
  ax2 = ax1.twinx()
  ax1.set_ylim([0, 65])
  ax2.set_ylim([0, 65])

  # Gerando o gráfico de inclinação (pontos + curva + texto)
  for i in range(5):
    # Adicionando o ponto inicial e final
    ax1.scatter(x=df_inc.index, y=df_inc.iloc[:, i], color=cores[i], s=50)

    # Adicionando a inclinação
    ax1.plot(df_inc.index, df_inc.iloc[:, i], color=cores[i], lw=3)

    # Adicionando as regiões no label
    ax1.text(x=2022.5, y=df_inc.iloc[-1, i], s=df_inc.columns[i], fontsize=15, fontweight='bold', color=cores[i], va='center')

  ## Personalizando o gráfico
  ax1.set_title('Participação do PIB relativo à indústria das regiões do Brasil', fontsize = 18, color = CINZA1, loc='left', pad = 40)
  ax1.text(0, 1.05, "Entre os anos de 2010, 2015 e 2020 (em %)", transform=ax1.transAxes, color= CINZA3, fontsize=12, ha='left', va='center')

  return ax1, ax2, cores

ax1, ax2, cores = grafico_slope_pib()
```

```python
# Gerando o gráfico de inclinação
ax1, ax2, cores = grafico_slope_pib()

# Retirando os ticks do eixo y de ambos os lados e as bordas exeto a da base
ax1.tick_params(left = False)
ax1.spines[['top', 'left', 'right']].set_visible(False)
ax2.tick_params(right=False)
ax2.set_frame_on(False)

# Ajustando as propriedades do eixo y da esquerda
plt.setp(ax1, xticks=[2010, 2015, 2020], xticklabels=['2010', '2015', '2020'], yticks=df_inc.iloc[0, :],
         yticklabels=[f'{str(valor)}%' if i!= 4 else f'{str(valor)}%\n\n' for i, valor in enumerate(df_inc.iloc[0, :])])

# Ajustando as propriedades dos pontos centrais
valores = df_inc.iloc[1, :].values
y_pos = [53, 17, 11, 4, -1]
for i in range(len(valores)):
  ax1.text(2015, y_pos[i], f'{valores[i]}%', weight='bold', fontsize=14, color=cores[i], ha='center', transform=ax.transData)

# Ajustando as propriedades do eixo y da direita
plt.setp(ax2, yticks=df_inc.iloc[2, :], yticklabels=[f'{str(valor)}%' for i, valor in enumerate(df_inc.iloc[2,:])])

# Ajustando propriedades dos dois eixos y
for ax in (ax1, ax2):
  for i, item in enumerate(ax.get_yticklabels()):
    item.set_fontsize(14)
    item.set_fontweight('bold')
    item.set_color(cores[i])

# Ajustando propriedades do eixo x
ax1.xaxis.set_tick_params(labelsize=14, labelcolor=CINZA3)
ax1.spines['bottom'].set_color(CINZA3)

plt.show()
```
